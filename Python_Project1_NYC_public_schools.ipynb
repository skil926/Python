```python
import pandas as pd 
import numpy as np
import matplotlib. pyplot as plt
```

```python
#display data set to analyze
df = pd.read_csv ('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
df
```

```python
#cleaning task
df.replace('s','4',inplace = True)
df.describe
```

```python
#drop columns with null/NaN values
df=df.replace({'na':np.NaN})
df=df.replace({'0':np.NaN})
df_n=df.dropna(axis=1)
df_n
```

```python
#subset containing only High Schools in District 26
df_ds26 = df_n['School DBN'].str.contains('26Q') & (df_n['School Level'] == 'High school')
df_n[df_ds26]
```

```python
#sort District 26 high schools by total tested to see lowest and highest tested school and year 
df_n[df_ds26].groupby(['School Name','Year'])['Total Tested'].count().sort_values(ascending=False)
```

```python
# select 1 high school (Bayside HS) in district 26 to compare against all high schools in district 26.
df_bhs = df_n['School DBN'].str.contains('26Q') & (df_n['School Level'] == 'High school') & (df_n['School Name']=='Bayside High School')
df_n[df_bhs].groupby(['Year'])['Total Tested'].count().sort_values(ascending = False)
```

```python
df_n[df_ds26].groupby(['Year','School Name'])['Total Tested'].count().sort_values(ascending = False).plot(kind='bar')
```
